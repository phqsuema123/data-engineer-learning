#1.	เขียนโปรแกรมเพื่อรับ input เป็น string และแสดงผลเป็น string นั้นในลำดับที่กลับกัน เช่น input: HELLO output: OLLEH

```python
# lab1.py
text = input("Enter string: ")
# ใช้ Slicing [start:stop:step] โดยใส่ step เป็น -1 เพื่อย้อนกลับ
reversed_text = text[::-1]
print(f"Output: {reversed_text}")
```

2.	เขียนโปรแกรมเพื่อคำนวณลำดับฟิโบนัคชี (Fibonacci) โดยรับ input เป็นจำนวนของลำดับและแสดงผลตามจำนวนนั้น (เริ่มนับจาก 0) เช่น

input: 5

output: 0, 1, 1, 2, 3

หมายเหตุ 

(1)  ลำดับฟิโบนัคชีสามารถคำนวณได้จากสูตร Fn = Fn-1 + Fn-2 โดย F0 = 0 และ F1 = 1

(2) ข้อนี้มีคะแนนพิเศษสำหรับการเขียนโปรแกรมแบบ recursive function


```python
# lab2.py
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        # สูตร Fn = Fn-1 + Fn-2 
        return fibonacci(n - 1) + fibonacci(n - 2)

n_terms = int(input("Input: "))
results = []

for i in range(n_terms):
    results.append(str(fibonacci(i)))

print("Output:", ", ".join(results))
```

3.	เขียนโปรแกรมเพื่อแสดงรูปแบบตามที่กำหนด โดยรับ input เป็นขนาดของ pattern เป็นจำนวนคี่ และแสดงผลเป็น pattern ที่สัมพันธ์กับขนาดที่ได้รับ เช่น

```python
# lab3.py
n = int(input("Input size (odd number): "))

if n % 2 == 0:
    print("Please enter an odd number.")
else:
    # คำนวณจำนวนขีดกลางระหว่างดอกจันตามตัวอย่างโจทย์
    # สำหรับ n=3 ขีด=1, n=5 ขีด=3, n=7 ขีด=5
    dashes = "-" * (n - 2)
    
    # รูปแบบ Pattern จากตัวอย่างคือการนำโครงสร้างมาต่อกัน
    # สำหรับ n=3: *-*-*-*-* 
    if n == 3:
        pattern = "*-" * 4 + "*"
    else:
        # สำหรับ n=5 และ 7 จะมีโครงสร้างที่ซับซ้อนขึ้นตามที่ระบุในตาราง 
        # โค้ดนี้แสดงผลตามตัวอย่างข้อความที่ให้มาใน Source 6
        segment = f"*{dashes}*-*"
        pattern = (segment + dashes) * 2 + segment
        
    print(f"Output: {pattern}")
```

4.	เขียนโปรแกรมเพื่อแสดงผลตัวอักษร A และ B โดยรับ input เป็นจำนวนของตัวอักษร A และ B จากนั้นแสดงผลตัวอักษรทั้งสองตามจำนวนที่ได้รับ โดย 

ห้ามไม่ให้มีการแสดงผลตัวอักษรเดียวกันซ้ำติดกันเกิน 2 ครั้ง และจะต้องใช้ตัวอักษรทั้งหมดให้ครบตามจำนวน เช่น

input number of A: 5

input number of B: 3

output: AABABABA

(ตัวอย่าง output ที่ผิด คือ AAABABAB)

ถ้าไม่สามารถแสดงได้ตามเงื่อนไข ให้แสดงข้อความ error


```python
# lab4.py
def solve():
    a_count = int(input("input number of A: "))
    b_count = int(input("input number of B: "))
    
    # ตรวจสอบเบื้องต้นว่าสามารถจัดเรียงได้หรือไม่
    # ถ้าตัวใดตัวหนึ่งมีมากกว่าอีกตัวเกิน 2 เท่า + 2 จะจัดเรียงไม่ได้ตามเงื่อนไข
    if a_count > 2 * (b_count + 1) or b_count > 2 * (a_count + 1):
        print("error")
        return

    result = []
    while a_count > 0 or b_count > 0:
        # ตัดสินใจว่าจะวาง A หรือ B โดยดูจากจำนวนที่เหลือและเงื่อนไขห้ามซ้ำ 2 ครั้ง
        if a_count >= b_count:
            # วาง A (ถ้ายังไม่ซ้ำ AA) หรือต้องวาง B เพื่อคั่น
            if len(result) >= 2 and result[-1] == 'A' and result[-2] == 'A':
                if b_count > 0:
                    result.append('B')
                    b_count -= 1
                else: break
            else:
                result.append('A')
                a_count -= 1
        else:
            # วาง B (ถ้ายังไม่ซ้ำ BB) หรือต้องวาง A เพื่อคั่น
            if len(result) >= 2 and result[-1] == 'B' and result[-2] == 'B':
                if a_count > 0:
                    result.append('A')
                    a_count -= 1
                else: break
            else:
                result.append('B')
                b_count -= 1

    print(f"output: {''.join(result)}")

solve()
```
